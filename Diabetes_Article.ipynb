```python
a1=data['Age'].values
b1=data['Glucose'].values
X=np.array(list(zip(a1,b1)))
plt.scatter(a1,b1, c='red')
```

```python
from scipy.cluster.hierarchy import fcluster, linkage
```

```python
"""
linkage matrix formation which can be passed to fcluster to form clusters based on distances
ward is chosen as the metric to get the minimum variance clusters
"""
Z = linkage(X, method='ward') 

"""
based on the linkage matrix and the given threshold, 
forms clusters and puts label to each datapoint
"""
res = fcluster(Z, 350, criterion='distance' )
```

```python
data[res==1].describe()
```

```python
data[res==2].describe()
```
